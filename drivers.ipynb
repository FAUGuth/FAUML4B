```python
import streamlit as st
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Page setup
st.set_page_config(page_title="F1 World Champions", layout="centered")
st.title("üèÅ F1 World Champions ‚Äì Number of Titles")

# Sample data
data = {
    "Driver": [
        "Lewis Hamilton",
        "Michael Schumacher",
        "Sebastian Vettel",
        "Ayrton Senna",
        "Fernando Alonso",
        "Max Verstappen",
        "Niki Lauda",
        "Alain Prost"
    ],
    "Titles": [7, 7, 4, 3, 2, 3, 3, 4]
}

df = pd.DataFrame(data)

# Sort data for plotting
df_sorted = df.sort_values("Titles", ascending=True)

# Plotting with Seaborn
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(
    data=df_sorted,
    x="Titles", y="Driver",
    palette="coolwarm",
    ax=ax
)

# Add value labels to bars
for i, val in enumerate(df_sorted["Titles"]):
    ax.text(val + 0.1, i, str(val), color='black', va='center')

# Chart titles and labels
ax.set_title("F1 World Championships by Driver")
ax.set_xlabel("Number of Titles")
ax.set_ylabel("Driver")
st.pyplot(fig)
```
